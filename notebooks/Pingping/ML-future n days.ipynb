```python
from tvDatafeed import TvDatafeed, Interval
```

```python
tv = TvDatafeed()
```

```python
n=1
```

```python
eth=tv.get_hist(symbol='ETHUSDT',exchange='BINANCE',interval=Interval.in_daily,n_bars=5000)
eth=eth.reset_index()
eth2=eth[eth["datetime"]>"2020-01-01"].copy()
eth2
```

```python
bnb=tv.get_hist(symbol='BNBUSDT',exchange='BINANCE',interval=Interval.in_daily,n_bars=5000)
bnb=bnb.reset_index()
bnb2=bnb[bnb["datetime"]>"2020-01-01"].copy()
bnb2
```

```python
ada=tv.get_hist(symbol='ADAUSDT',exchange='BINANCE',interval=Interval.in_daily,n_bars=5000)
ada=ada.reset_index()
ada2=ada[ada["datetime"]>"2020-01-01"].copy()
ada2
```

```python
bitcoi = tv.get_hist(symbol='BTCUSDT', exchange='BINANCE', interval=Interval.in_daily, n_bars=5000)
bitcoi = bitcoi.reset_index()
# dataTypeSeries = bitcoi.dtypes
# print('Data type of each column of Dataframe :')
# print(dataTypeSeries)
bitcoi2=bitcoi[bitcoi["datetime"]>"2020-01-01"].copy()
bitcoi2=bitcoi2[["datetime","volume","open","high","low","close"]]
bitcoi2
```

```python
bitcoin=bitcoi2.merge(eth2[["datetime","close"]],on='datetime',how='left',suffixes=('_bitcoin', '_eth'))
bitcoin=bitcoin.merge(bnb2[["datetime","close"]],on='datetime',how='left')
bitcoin=bitcoin.merge(ada2[["datetime","close"]],on='datetime',how='left',suffixes=('_bnb', '_ada'))
bitcoin
```

```python
plt.figure(figsize = (12, 7))
plt.plot(bitcoin["datetime"], bitcoin["close_bitcoin"], color='goldenrod', lw=2)
plt.title("Bitcoin Price over time", size=25)
plt.xlabel("Time", size=20)
plt.ylabel("$ Price", size=20)
plt.show()
```

```python
plt.figure(figsize = (12, 7))
plt.plot(bitcoin["datetime"], bitcoin["volume"], color='royalblue', lw=2)
plt.title("Bitcoin Volume over time", size=25)
plt.xlabel("Time", size=20)
plt.ylabel("Volume", size=20);
```

```python
bitcoin.isnull().sum()
```

```python
required_features = ['volume','open', 'high', 'low','close_eth','close_bnb','close_ada']
output_label = 'close_bitcoin'
```

```python
x_train, x_test, y_train, y_test = train_test_split(bitcoin[required_features],bitcoin[output_label],test_size = 0.2)
```

```python
x_test
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
future_set = bitcoin.tail(n).copy()
future_set
```

```python
prediction = model.predict(future_set[required_features])
prediction
```

```python
import datetime
future_set["datetime"]=future_set["datetime"]+datetime.timedelta(days=n)
future_set["predict"]=prediction
future_set.reset_index()
print("The prediction of bitcoin price on",future_set["datetime"].iloc[0],"is",future_set["predict"].iloc[0])
```
