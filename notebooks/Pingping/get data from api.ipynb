```python
import requests
import json
import pandas as pd
response_API = requests.get('https://api.blockchain.info/charts/n-transactions?timespan=100years&rollingAverage=24hours&format=json&sampled=false')
#print(response_API.status_code)
```

```python
data = response_API.text
```

```python
parse_json=json.loads(data)
```

```python
parse_json
```

```python
df = pd.json_normalize(parse_json['values'])
```
