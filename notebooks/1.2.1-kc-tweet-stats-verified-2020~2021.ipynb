```python
# import packages
import twint
import nest_asyncio
from lib import bittwint
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

nest_asyncio.apply()
```

```python
# scraping function
df = bittwint.scrape_twitter(
    "bitcoin", 
    100000, 
    "2021-01-14", 
    verified=True, 
    output="../data/raw/bitcoin_verified_tweets_5.csv"
)

display(df)
```

```python
# get all scraped data
df1 = pd.read_csv("../data/raw/bitcoin_verified_tweets_1.csv", low_memory=False)
df2 = pd.read_csv("../data/raw/bitcoin_verified_tweets_2.csv", low_memory=False)
df3 = pd.read_csv("../data/raw/bitcoin_verified_tweets_3.csv", low_memory=False)
df4 = pd.read_csv("../data/raw/bitcoin_verified_tweets_4.csv", low_memory=False)
df5 = pd.read_csv("../data/raw/bitcoin_verified_tweets_5.csv", low_memory=False)
```

```python
# concat all data
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

df_id = df.drop_duplicates(subset=['id'])

display(df_id.head())
```

```python
# pd.set_option('display.max_columns', None)

df_id[['video']].value_counts()
# df_id
```

```python
df_id_col = df_id.drop(columns=['date', 'time', 'timezone', \
                             'place', 'retweet', 'near', 'geo', 'source', \
                             'user_rt_id', 'user_rt', 'retweet_id', 'retweet_date', \
                            'translate', 'trans_src', 'trans_dest'])

display(df_id_col.head())
```

```python
# change data types
df_id_col["created_at"] = pd.to_datetime(df_id_col["created_at"])
df_id_col = df_id_col.astype({
    'username': 'string',
    'name': 'string',
    'tweet': 'string',
    'language': 'string',
    'link': 'string',
    'quote_url': 'string',
    'video': 'bool',
    'thumbnail': 'string',
})

display(df_id_col.dtypes)
```

```python
# filer date 312521
df_id_col_date = df_id_col.set_index('created_at').sort_index()

df_id_col_date = df_id_col_date.loc['2020':'2021']
df_id_col_date = df_id_col_date.reset_index()

display(df_id_col_date)
```

```python
# export to pickle
df_id_col_date.to_pickle("../data/interim/tweets_verified_2020-2021.pkl")
```
